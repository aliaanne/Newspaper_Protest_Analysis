```python
new_dict = dict([(value, key) for key, value in country_list])

dict_countries = {}

dict_countries['na'] = 0
for country in new_dict:
    dict_countries[country] = 0

lower_dict_countries =  {k.lower(): v for k, v in dict_countries.items()}

print(lower_dict_countries)
```

### Create for loop to assign each article a country

Set a counter equal to zero. 

For each row in the dataset the temporary dictionary will be your lowercase country dictionary and the content will be your article content. 

For each country in the dictionary of countries, sum the amount of times it shows up in the article using a regular expression and assign it to a variable called count. 

Generate a variable called country_max, which is the country with the highest count value for a given article. 

Print the name of that country in the dataframe column named country at the appropriate indexed row. 

Print the number of rows left to process with a counter to keep track of progress.

```python
counter = 0
for index, row in dataset.iterrows(): 
    dict_temp = lower_dict_countries
    content = row['content']
   

    for country in lower_dict_countries:
        count = sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(country), content))
        #print(count)
        dict_temp[country] = count

    #print(dict_temp)
    country_max = max(dict_temp.items(), key=operator.itemgetter(1))[0]
    #print(country_max)
    
    dataset.at[index, 'country'] = country_max
    
    print('number of rows left to process', dataset.shape[0] - counter, '/', dataset.shape[0])
    counter = counter + 1
    
dataset.head()
```

### Set up anger dictionary

Create a new empty column in the dataset called anger.

Import a csv of anger words (created by me) into a dataframe with column for the emotion and an empty column for the value. 

Turn the dataframe into a dictionary and view it. 

```python
dataset['anger'] = ''

angerdata = pd.read_csv("/Users/aliabraley/Desktop/anger_dictionary.csv", header = None, names = ['emotion', 'value'])

anger_dict = dict(zip(angerdata.emotion, angerdata.value))

anger_dict
```

### Run for loops to count anger words

#### External for loop

Set a counter equal to zero.

For each row in the dataset the temporary dictionary will be your anger dictionary and the content will be your article content.

Create an emotion counter variable, counter_emotion and set equal to zero. 

#### Internal for loop

For each emotion in the dictionary of emotions, sum the amount of times it shows up in the article using a regular expression and assign it to a variable called count.

Add the sum of that emotion to the sum of the previously iterated emotions to generate a count of all of the anger words in a given article.

Assign that total number to the column anger in the dataset at the appropriate indexed row. 

Print the number of rows left to process with a counter to keep track of progress.



```python
counter = 0
for index, row in dataset.iterrows(): 
    dict_temp = anger_dict
    content = row['content']
   
    counter_emotion = 0
    for emotion in dict_temp:
        count = sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(emotion), content))
        dict_temp[emotion] = count
        counter_emotion = counter_emotion + count 
    
    dataset.at[index, 'anger'] = counter_emotion
    
    print('number of rows left to process', dataset.shape[0] - counter, '/', dataset.shape[0])
    counter = counter + 1
    
dataset.head()
```
